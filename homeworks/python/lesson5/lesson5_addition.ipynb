```python
# пишу функцию на + "id" и применяю ко всему столбцу
```

```python
def add_id (x):
    if x.startswith('id'):
        return x
    else:
        return "id" + x
```

```python
df['SUBS_ID'] = df['SUBS_ID'].apply(add_id)
```

```python
# ищу разницу по заданию: продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
```

```python
prod_activations_logs
```

```python
prod_activations_logs['START_DTTM'] = pd.to_datetime(prod_activations_logs['START_DTTM'], format='%d-%m-%Y %H:%M')
prod_activations_logs['END_DTTM'] = pd.to_datetime(prod_activations_logs['END_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
# format='%d-%m-%Y %H:%M' позволяет корректно перевести дату иначе бы месяц мог стать днём, а день месяцем
```

```python
# я не понял почему, но следующий способ вызвал ошибку (поэтоу использвоал способ по-старинке) prod_activations_logs = prod_activations_logs.assign(difference = prod_activations_logs.END_DTTM + prod_activations_logs.START_DTTM)
```

```python
prod_activations_logs['difference'] = prod_activations_logs['END_DTTM'] - prod_activations_logs['START_DTTM']
```

```python
prod_activations_logs
```

```python
# сравниваю значения в колонке  difference c 5 минутами, есть лбопытный способ:
```

```python
prod_activations_logs['timedelta_column'] = prod_activations_logs.difference > '5m'
```
