```python
# 8 Вернемся к анализу полного датасета retail. Создайте колонку Revenue с суммой покупки, используя колонки Quantity и UnitPrice.
```

```python
retail = retail.assign(Revenue = retail.Quantity * retail.UnitPrice)
```

```python
retail
```

```python
# 9 Для каждой транзакции (InvoiceNo), посчитайте финальную сумму заказа. 
# В качестве ответа укажите топ-5 (InvoiceNo) по сумме заказа (через запятую с пробелом и в порядке убывания TotalRevenue)
```

```python
k = retail.groupby('InvoiceNo', as_index = False).agg(TotalRevenue=('Revenue', 'sum')).sort_values('TotalRevenue', ascending = False).head()
```
